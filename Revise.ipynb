```python
# Prime Number
n=19
count=0
for i in range(1,n+1):
    if n%i==0:
        count+=1
if count==2:
    print("prime")
else:
    print("not prime")
```

```python
n=19
is_prime=False

if n>=1:
    is_prime=True
else:
    i=2
    while i>n:
        if n%i==0:
            is_prime=False
            break
        i+=1
print(is_prime)
```

```python
# Palindrome Number 
n=100
digits=n
rev=0
while digits>0:
    d=digits%10
    rev=rev*10+d
    digits=digits//10
if n==rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
n=100
n_1=str(n)
if n_1==n_1[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Check Armstrong Number
n="159"
power=len(n)
sum=0
for i in n:
    sum=sum+int(i)**power
if int(n)==sum:
    print("Armstrong")
else:
    print("Not")
```

```python
# Print Fibonacci Series
n=8
a=0
b=1
for i in range(n):
    print(a)
    c=a+b
    a=b
    b=c
```

```python
# Find Factorial of a number
n=5
fact=1
for i in range(1,n+1):
    fact*=i
print(fact)
```

```python
# Reverse a number
n=123
n_1=str(n)
rev=""
for i in n_1:
    rev=i+rev
print(rev)
    
```
