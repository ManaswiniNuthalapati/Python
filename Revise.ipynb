```python
# Prime Number
n=19
count=0
for i in range(1,n+1):
    if n%i==0:
        count+=1
if count==2:
    print("prime")
else:
    print("not prime")
```

```python
n=19
is_prime=False

if n>=1:
    is_prime=True
else:
    i=2
    while i>n:
        if n%i==0:
            is_prime=False
            break
        i+=1
print(is_prime)
```

```python
# Palindrome Number 
n=100
digits=n
rev=0
while digits>0:
    d=digits%10
    rev=rev*10+d
    digits=digits//10
if n==rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```
