```python
# Prime Number
n=19
count=0
for i in range(1,n+1):
    if n%i==0:
        count+=1
if count==2:
    print("prime")
else:
    print("not prime")
```

```python
n=19
is_prime=False

if n>=1:
    is_prime=True
else:
    i=2
    while i>n:
        if n%i==0:
            is_prime=False
            break
        i+=1
print(is_prime)
```

```python
# Palindrome Number 
n=100
digits=n
rev=0
while digits>0:
    d=digits%10
    rev=rev*10+d
    digits=digits//10
if n==rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
n=100
n_1=str(n)
if n_1==n_1[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Check Armstrong Number
n="159"
power=len(n)
sum=0
for i in n:
    sum=sum+int(i)**power
if int(n)==sum:
    print("Armstrong")
else:
    print("Not")
```

```python
# Find GCD and LCM
a=12
b=18
if a<b:
    smaller=a
else:
    smaller=b
gcd=1
for i in range(1,smaller+1):
    if a%i==0 and b%i==0:
        gcd=i
print(gcd)
lcm=(a*b)//gcd
print(lcm)
    
```

```python
# LCM Number
a=12
b=18
if a>b:
    big=a
else:
    big=b
while True:
    if big%a==0 and big%b==0:
        print(big)
        break
    big+=1
```

```python
# Check Perfect Number
n=7
li=[]
for i in range(1,n):
  if n%i==0:
    li.append(i)
sum=0
for i in li:
  sum+=i
if n==sum:
  print("Perfect Number")
else:
  print("Not")
```

```python
# find duplicate characters
a="madam"
freq={}
for i in a:
  if i in freq:
    freq[i]+=1
  else:
    freq[i]=1
print(freq)
for key,value in freq.items():
  if value>1:
    print(key,end=" ")
```

```python
# Check Anagram
a="listen"
b="silent"
if len(a)!=len(b):
  print(False)
for i in a:
  if a.count(i)!=b.count(i):
    print(False)
else:
  print(True)
```
