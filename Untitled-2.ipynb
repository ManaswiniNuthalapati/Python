```python
# Find the difference between the largest and smallest element in a tuple.

a=(4,3,9,5,8,6,2)
large=a[0]
small=a[0]
for i in a:
    if i>large:
        large=i
    if i<small:
        small=i
print(large)
print(small)
print("Difference:",large-small)

    
```

```python
# Find the **second smallest** element in a tuple.

a=(5,4,3,8,7)
b=list(a)
small=b[0]
sec=b[0]
for i in b:
    if i<small:
        sec=small
        small=i
    elif i<sec and i!=small:
        sec=i
print(small)
print(sec)
        
    
```

```python
# left rotate tuple
a=(1,2,3,4)
b=1
b=b%len(a)
res=a[b:]+a[:b]
print(res)
```

```python
# right rotate tuple
a=(1,2,3,4)
b=1
b=b%len(a)
res=a[-b:]+a[:-b]
print(res)
```
